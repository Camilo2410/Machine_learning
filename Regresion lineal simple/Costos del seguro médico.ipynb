## importamos las bibliotecas necesarias.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

## Cargamos el conjunto de datos

```python
df = pd.read_csv('dataset.csv')
```

## Visualizamos las primeras filas del conjunto de datos para entender su estructura

```python
print(df.head())
```

## Separacion de Variables

```python
X = df[['age']]  # Usamos doble corchete para mantener X como un DataFrame
y = df['charges']
```

##  Division del conjunto de datos

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Creamos el modelo de regresión lineal

```python
model = LinearRegression()
```

## Entrenamos el modelo utilizando el conjunto de entrenamiento

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Error cuadrático medio (MSE): {mse}")
print(f"Coeficiente de determinación (R^2): {r2}")
```

```python
# Visualización de la línea de regresión sobre los datos de prueba
```

```python
plt.scatter(X_test, y_test, color='black', label='Datos Reales')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Línea de Regresión')
plt.xlabel('Edad')
plt.ylabel('Cargos del Seguro')
plt.title('Regresión Lineal Simple de Edad vs Cargos del Seguro')
plt.legend()
plt.show()
```

El valor de R^2 es 0.125, lo que indica que aproximadamente el 12.5% de la variabilidad en los cargos del seguro puede ser explicado por la edad de los asegurados. Esto sugiere que la edad tiene un efecto moderado en los costos del seguro en este conjunto de datos, pero también implica que hay otros factores no incluidos en este modelo simple que tienen un impacto significativo en los cargos del seguro.

Utilizar regresion lineal simple no es la mejor opcion para solucionar este problema pero es un ejemplo de aplicacion.
